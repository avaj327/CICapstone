```python
import requests,json,os
import pandas as pd
import info
API_KEY = info.CENSUS_KEY
print(API_KEY)
```

```python
variable_dict = {
    "B07009_005E":"Individuals with Bachelor's Degrees",
    "B07010_047E":"# of People who Moved here with Income",
    "B07401_001E":"Total People Living in this Area 1 Year Ago",
    "B25003_003E":"# Houses Rented",
    "B25118_015E":"# Houses Rented for <$5,000",
    "B08124_002E":"Higher Education Occupuations"
}
```

```python
chosen_variables = "NAME"
for field in variable_dict:
    chosen_variables += f",{field}"
print(chosen_variables)
```

```python
#Define all the parts of the query url
# = "NAME,B24134_186E,B24134_226E,B25003_003E,B25118_015E,B08124_002E"
host = "https://api.census.gov/data"
#host_variables = "https://api.census.gov/data/2020/acs/acs5/variables"
year = "/2020"
dataset_acronym = "/acs/acs5"
g = "?get="
#variables = chosen_variables
variables = chosen_variables
#location = '&for=state:*'
user_key = f"&key={API_KEY}"
```

```python
city_list = ["DC","New York","Philadelphia","Detroit","Billings",
             "Minneapolis","Los Angeles","Portland","Seattle","Phoenix","Houston","Atlanta"]
city_info = [["53",  "63000"],
 ["48",  "35000"],
 ["42",  "60000"],
 ["41",  "59000"],
 ["36",  "51000"],
 ["30",  "06550"],
 ["04",  "55000"],
 ["06",  "44000"],
 ["13",  "04000"],
 ["26",  "22000"],
 ["27",  "43000"],
 ["11",  "50000"]]
# District of Columbia","705749","11","50000"
city_dict = {"63000":"Seattle",
             "35000":"Houston",
             "60000":"Philadelphia",
             "59000":"Portland",
             "51000":"New York",
             "06550":"Billings",
             "55000":"Phoenix",
             "44000":"Los Angeles",
             "04000":"Atlanta",
             "22000":"Detroit",
             "43000":"Minneapolis",
             "50000":"Washington"}
```

```python
results = []
for state,code in city_info:
    location = f'&for=place:{code}&in=state:{state}'
    query_url = f"{host}{year}{dataset_acronym}{g}{variables}{location}{user_key}"
    response = requests.get(query_url)
    json_data = response.json()
    city_df = pd.DataFrame(json_data, columns=json_data[0]).drop(0).rename(columns=variable_dict).reset_index().drop(columns='index')
    city_df['city'] = city_dict[code].lower()
    results.append(city_df)
response_df = pd.concat(results)
display(response_df)
```
